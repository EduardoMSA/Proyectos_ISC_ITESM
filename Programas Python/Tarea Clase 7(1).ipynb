```python
#Eduardo Méndez A01632947

import random
import time
import matplotlib.pyplot as plt

#1.- La complejidad del algoritmo Insertion Sort cuadno se tiene
#una lista aleatoria, es decir, en el caso promedio es de: O(n^2)

#2.-

def selectionSort(lista):
    for i in range(len(lista)-1):
        max=i
        for j in range(i+1,len(lista)):
            if(lista[j]<lista[max]):
                max = j
        tmp = lista[i]
        lista[i] = lista[max]
        lista[max] = tmp
        
def insertionSort(lista):
    for i in range(1, len(lista)):
        aux = lista[i]
        j = i - 1
        while j>=0 and aux < lista[j]:
            lista[j+1] = lista[j]
            j-=1
        lista[j+1]=aux
        
def bubbleSort(lista):
    for i in range(len(lista)):
        for j in range(0, len(lista)-i-1):
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]
                
def shellSort(lista):
    mitad = len(lista)//2
    while mitad > 0:
        for i in range(mitad, len(lista)):
            aux = lista[i]
            j = i
            while j >= mitad and lista[j-mitad] > aux:
                lista[j] = lista[j-mitad]
                j -= mitad
            lista[j]=aux
        mitad//=2
        
timepos1 = []
timepos2 = []
timepos3 = []
timepos4 = []
        
lista1 = random.sample(range(100), 10)
lista2 = random.sample(range(100), 20)
lista3 = random.sample(range(100), 30)
lista4 = random.sample(range(100), 40)
lista5 = random.sample(range(100), 50)

tmp = lista1
start = time.time()
selectionSort(tmp)
end = time.time()
timepos1.append(end-start)

tmp = lista2
start = time.time()
selectionSort(tmp)
end = time.time()
timepos1.append(end-start)

tmp = lista3
start = time.time()
selectionSort(tmp)
end = time.time()
timepos1.append(end-start)

tmp = lista4
start = time.time()
selectionSort(tmp)
end = time.time()
timepos1.append(end-start)

tmp = lista5
start = time.time()
selectionSort(tmp)
end = time.time()
timepos1.append(end-start)



tmp = lista1
start = time.time()
insertionSort(tmp)
end = time.time()
timepos2.append(end-start)

tmp = lista2
start = time.time()
insertionSort(tmp)
end = time.time()
timepos2.append(end-start)

tmp = lista3
start = time.time()
insertionSort(tmp)
end = time.time()
timepos2.append(end-start)

tmp = lista4
start = time.time()
insertionSort(tmp)
end = time.time()
timepos2.append(end-start)

tmp = lista5
start = time.time()
insertionSort(tmp)
end = time.time()
timepos2.append(end-start)



tmp = lista1
start = time.time()
bubbleSort(tmp)
end = time.time()
timepos3.append(end-start)

tmp = lista2
start = time.time()
bubbleSort(tmp)
end = time.time()
timepos3.append(end-start)

tmp = lista3
start = time.time()
bubbleSort(tmp)
end = time.time()
timepos3.append(end-start)

tmp = lista4
start = time.time()
bubbleSort(tmp)
end = time.time()
timepos3.append(end-start)

tmp = lista5
start = time.time()
bubbleSort(tmp)
end = time.time()
timepos3.append(end-start)



tmp = lista1
start = time.time()
shellSort(tmp)
end = time.time()
timepos4.append(end-start)

tmp = lista2
start = time.time()
shellSort(tmp)
end = time.time()
timepos4.append(end-start)

tmp = lista3
start = time.time()
shellSort(tmp)
end = time.time()
timepos4.append(end-start)

tmp = lista4
start = time.time()
shellSort(tmp)
end = time.time()
timepos4.append(end-start)

tmp = lista5
start = time.time()
shellSort(tmp)
end = time.time()
timepos4.append(end-start)

plt.plot(timepos1, label="SelectionSort")
plt.plot(timepos2, label="InsertionSort")
plt.plot(timepos3, label="BubbleSort")
plt.plot(timepos4, label="ShellSort")
plt.title("Tiempo transcurrido por tamaño de lista")
plt.legend(loc = "upper left")
plt.ion
plt.show


```
