```python
#Georgina Gonzalez A01632886
#Eduardo Méndez A01632947
#Sebastián Serna A01633102

import time
import matplotlib.pyplot as plt

#Ejercicio 1
#Complejidad O(1)
def primerValor(lista):
    print(lista[0])
    
#Ejercicio 2
#Complejidad O(n)
def imprimeLista(lista):
    for element in lista:
        print(element, end=" ")
    print()
    
#Ejercicio 3    
#Complejidad O(n^2)
def pares(num):
    #print("Ejercicio 3")
    for x in range (0, len(num)):
        for y in range(0, len(num)):
            print(num[x] , "," , num[y])
    print()


#Ejercicio 4
#Complejidad O(n/2)
def completeness(parentesis):
    i = 0
    j = len(parentesis)-1
    while (i<=j):
        if parentesis[i] == "(" and parentesis[j] != ")":
            print("False")
            return False
        elif parentesis[i] == "[" and parentesis[j] != "]":
            print("False")
            return False
        elif parentesis[i] == "{" and  parentesis[j] != "}":
            print("False")
            return False
        elif parentesis[i] != "(" and parentesis[i] != "{" and parentesis[i] != "[":
            print("False")
            return False
        i = i + 1
        j = j - 1
    print("True")
    return True
            

#Ejercicio 5
#Complejidad O(nlogn)
def isAnagram(str1, str2):
    str1 = str1.replace(" ","")
    str2 = str2.replace(" ","")
    str1 = str1.lower()
    str2 = str2.lower()
    if len(str1) != len(str2):
        print("False")
        return False
    for character in str1:
        length = len(str2)
        for character2 in str2:
            if character == character2:
                str2 = str2.replace(character2,"",1)
                break
        if length == len(str2):
            print("False")
            return False
    print("True")
    return True



lista = [0]
parentesis = "("
valoresE1 = []
valoresE2 = []
valoresE3 = []
valoresE4 = []

for n in range(0,29):
    
    start = time.time()
    primerValor(lista)
    end = time.time()
    valoresE1.append(end-start)
    
    start = time.time()
    imprimeLista(lista)
    end = time.time()
    valoresE2.append(end-start)
    
    start = time.time()
    pares(lista)
    end = time.time()
    valoresE3.append(end-start)
    
    start = time.time()
    completeness(parentesis)
    end = time.time()
    valoresE4.append(end-start)
    
    if(n%2 == 0):
        parentesis = parentesis + ")"
    else:
        parentesis = "(" + parentesis
    lista.append(n)

plt.plot(valoresE1, label="Ejercicio 1")
plt.plot(valoresE2, label="Ejercicio 2")
plt.plot(valoresE3, label="Ejercicio 3")
plt.plot(valoresE4, label="Ejercicio 4")
plt.title("Tiempo trenscurrido por tamaño de entrada")
plt.xlabel("Tamaño de entrada")
plt.ylabel("Tiempo")
plt.legend(loc="upper left")
plt.ion
plt.show
    
    



```
